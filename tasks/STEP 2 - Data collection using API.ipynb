## API request format


1920px-Request_format.svg.png

The response from the API is usually returned in a JSON or XML format. 


Heres an example of a JSON-formatted API response:

```
{
  "status": "OK",
  "result": {
    "id": 123,
    "name": "John Doe",
    "email": "john.doe@example.com"
  }
}
```

Heres an example of an XML-formatted API response:

```
<?xml version="1.0"?>
<response>
  <status>OK</status>
  <result>
    <id>123</id>
    <name>John Doe</name>
    <email>john.doe@example.com</email>
  </result>
</response>
```

## How to query API using Python

For that, we need Python library *requests*:

````python
import requests

url = 'https://api.spaceflightnewsapi.net/v3/articles?_limit=2'
response = requests.get(url)
response.json()
````

Usually, parameters are passed as JSON together with the request:

````python
base_url = 'https://api.spaceflightnewsapi.net/v3/'
endpoint = 'articles'
parameters = {'_limit': 5}

url = base_url + endpoint
response = requests.get(url, params=parameters)
response.json()
````

## Introducing Pandas

We will use Pandas to work with tabular data in Python. Pandas is using a two-dimensional data structures called DataFrames. 

````python
import pandas as pd
import json

df = pd.DataFrame(response.json()) # convert the response from the API to Pandas DataFrame
df
````

````python
# Save DataFrame to CSV
df.to_csv ('spaceflightnews.csv', index = None)
````

## The GDELT Project

The GDELT Project monitors the world's broadcast, print, and web news from nearly every corner of every country in over 100 languages and identifies the people, locations, organizations, themes, sources, emotions, counts, quotes, images and events.

The official website: https://www.gdeltproject.org/

There is an unofficial web interface for some GDELT's APIs, which we will use in the next tasks: https://gdelt.github.io/

Note that the web interface generates the API call for GDELT database using the parameters used for search.

For more options, full GDELT API documentation is available [here](https://blog.gdeltproject.org/gdelt-doc-2-0-api-debuts/)

<br>

For an example, we will use a search about heatwave in English language for the last day: https://api.gdeltproject.org/api/v2/doc/doc?query=heatwave%20sourcelang:eng&mode=ArtList&maxrecords=75&format=json&timespan=1d

So the JSON with API parameters will look like this:
```
{
  'query': 'heatwave sourcelang:eng',
  'mode': 'ArtList',
  'maxrecords': 75,
  'format': 'json',
  'timespan': '1d'
}
```

````python
# Get data from the API
base_url = 'https://api.gdeltproject.org/api/v2/'
endpoint = 'doc/doc'
parameters = {
  'query': 'heatwave sourcelang:eng',
  'mode': 'ArtList',
  'maxrecords': 75,
  'format': 'json',
  'timespan': '1d'
}

url = base_url + endpoint
response = requests.get(url, params=parameters)
response.json()
````

````python
# Convert response JSON to the DataFrame and save it as CSV
df = pd.DataFrame(response.json()['articles'])
df.to_csv ('gdelt_heatwave.csv', index = None)
````
